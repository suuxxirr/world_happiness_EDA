### Plotting

x 및 y 포인트 플로팅
plot() 함수는 다이어그램에서 점(마커)을 그리는 데 사용됩니다.

기본적으로 plot() 함수는 점에서 점으로 선을 그립니다.

이 함수는 다이어그램의 지점을 지정하기 위한 매개변수를 사용합니다.

매개변수 1은 x축의 점을 포함하는 배열입니다.

매개변수 2는 y축의 점을 포함하는 배열입니다.

(1, 3)에서 (8, 10)까지 선을 그려야 하는 경우 플롯 함수에 두 개의 배열 [1, 8] 및 [3, 10]을 전달해야 합니다.

```python

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

#### Plotting Without Line
마커만 플로팅하려면 '링'을 의미하는 단축 문자열 표기 매개변수 'o'를 사용할 수 있습니다.

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

#### Multiple Points
원하는 만큼 점을 그릴 수 있습니다. 두 축에 동일한 수의 점이 있는지 확인하십시오.

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

#### Default X-Points
x축에 점을 지정하지 않으면 y점의 길이에 따라 기본값 0, 1, 2, 3(등)을 갖게 됩니다.

따라서 위와 동일한 예를 취하고 x-점을 제외하면 다이어그램은 다음과 같이 표시됩니다.

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

### Matplotlib Labels and Title
#### Create Labels for a Plot
Pyplot에서는 xlabel() 및 ylabel() 함수를 사용하여 x축 및 y축에 대한 레이블을 설정할 수 있습니다.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

#### Create a Title for a Plot
Pyplot에서는 title() 함수를 사용하여 플롯의 제목을 설정할 수 있습니다.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

#### Set Font Properties for Title and Labels
xlabel(), ylabel() 및 title()에서 fontdict 매개 변수를 사용하여 제목 및 레이블의 글꼴 속성을 설정할 수 있습니다.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'AppleGothic','color':'blue','size':20}
font2 = {'family':'AppleGothic','color':'darkred','size':15}

# plt.title("Sports Watch Data", fontdict = font1)
plt.title("스포츠 관람 데이터", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

#### Position the Title
title()의 loc 매개변수를 사용하여 제목을 배치할 수 있습니다.

유효한 값은 'left', 'right' 및 'center'입니다. 기본값은 'center'입니다.


```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data", loc = 'left', fontsize = 20)
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.show()
```

### Grid
#### Add Grid Lines to a Plot

### Matplotlib Subplot
#### Display Multiple Plots
subplot() 함수를 사용하면 하나의 그림에 여러 플롯을 그릴 수 있습니다.

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

#1 row, 2 columns, the first plot.
plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

#1 row, 2 columns, the second plot.
plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

#### The subplot() Function
subplot() 함수는 그림의 레이아웃을 설명하는 세 개의 인수를 사용합니다.

레이아웃은 첫 번째와 두 번째 인수로 표시되는 행과 열로 구성됩니다.

세 번째 인수는 현재 플롯의 인덱스를 나타냅니다.

```python
plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.
```

```python
plt.subplot(1, 2, 2)
#the figure has 1 row, 2 columns, and this plot is the second plot.
```

따라서 2개의 행과 1개의 열이 있는 그림을 원하는 경우(두 플롯이 나란히 표시되지 않고 서로 위에 표시됨을 의미) 다음과 같은 구문을 작성할 수 있습니다.

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

하나의 그림에 원하는 만큼 플롯을 그릴 수 있습니다. 플롯의 행, 열 및 인덱스 수만 설명하면 됩니다.

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(x,y)

plt.show()
```

### Scatter
#### Creating Scatter Plots
Pyplot에서는 scatter() 함수를 사용하여 산점도를 그릴 수 있습니다.

scatter() 함수는 각 관찰에 대해 하나의 점을 그립니다. 길이가 같은 두 개의 배열이 필요합니다. 하나는 x축 값용이고 다른 하나는 y축 값용입니다.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

위 예의 관찰은 13대의 자동차가 지나간 결과입니다.

X축은 자동차의 나이를 나타냅니다.

Y축은 자동차가 통과할 때의 속도를 나타냅니다.

관찰 사이에 관계가 있습니까?

새 차일수록 더 빨리 달리는 것 같지만, 결국 우리가 등록한 차는 13대뿐이었기 때문에 우연의 일치일 수도 있습니다.

#### Compare Plots
위의 예에서 속도와 나이 사이에는 관계가 있는 것 같지만 다른 날의 관측값도 표시하면 어떻게 될까요?

산포도(Scatter Plot)가 우리에게 다른 것을 말해줄까요?

```python
#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

plt.show()
```

*두 플롯은 기본적으로 파란색과 주황색의 두 가지 다른 색상으로 표시됩니다. 이 장의 뒷부분에서 색상을 변경하는 방법을 배우게 됩니다.

두 플롯을 비교하면 둘 다 우리에게 같은 결론을 제공한다고 말하는 것이 안전하다고 생각합니다. 자동차가 새 차일수록 더 빨리 운전합니다.

#### Colors
color 또는 c 인수를 사용하여 각 산점도에 대해 고유한 색상을 설정할 수 있습니다.

```python
# 사용자 지정 색상
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = 'hotpink')

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = '#88c999')

plt.show()
```

#### Color Each Dot
c 인수의 값으로 색상 배열을 사용하여 각 점에 대해 특정 색상을 설정할 수도 있습니다.

*이를 위해 color 인수를 사용할 수 없으며 c 인수만 사용할 수 있습니다.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)

plt.show()
```

#### ColorMap
Matplotlib 모듈에는 여러 가지 사용 가능한 색상표가 있습니다.

컬러맵은 각 색상이 0에서 100 사이의 값을 갖는 색상 목록과 같습니다.

다음은 컬러맵의 예입니다.

이 컬러맵은 'viridis'라고 하며 보시다시피 0(보라색)부터 100(노란색)까지의 범위를 갖습니다.

#### How to Use the ColorMap
키워드 인수 cmap을 사용하여 컬러맵 값을 사용하여 컬러맵을 지정할 수 있습니다. 이 경우 Matplotlib에서 사용할 수 있는 내장 컬러맵 중 하나인 'viridis'입니다.

또한 산점도의 각 지점에 대해 하나의 값(0에서 100까지)이 있는 배열을 만들어야 합니다.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.show()
```

plt.colorbar() 문을 포함하여 그림에 컬러맵을 포함할 수 있습니다.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()

plt.show()
```

#### Size
s 인수로 점의 크기를 변경할 수 있습니다.

색상과 마찬가지로 크기 배열의 길이가 x축 및 y축 배열과 동일한지 확인하십시오.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)

plt.show()
```

#### Alpha
알파 인수를 사용하여 점의 투명도를 조정할 수 있습니다.

색상과 마찬가지로 크기 배열의 길이가 x축 및 y축 배열과 동일한지 확인하십시오.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```

#### Combine Color Size and Alpha
컬러맵을 다양한 크기의 도트와 결합할 수 있습니다. 도트가 투명한 경우 가장 잘 시각화됩니다.

### Bars
#### Creating Bars
Pyplot에서는 bar() 함수를 사용하여 막대 그래프를 그릴 수 있습니다.

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

bar() 함수는 막대의 레이아웃을 설명하는 인수를 사용합니다.

첫 번째 및 두 번째 인수가 배열로 나타내는 범주 및 해당 값입니다.

```python
x = ["APPLES", "BANANAS"]
y = [400, 350]
plt.bar(x, y)
```

#### Horizontal Bars
막대를 수직이 아닌 수평으로 표시하려면 barh() 함수를 사용하십시오.

```python
# 4 horizontal bars
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

#### Bar Color
bar() 및 barh()는 키워드 인수 color를 사용하여 막대의 색상을 설정합니다.

```python
# 4 red bars
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

# plt.bar(x, y, color = "#fccf03")
plt.bar(x, y, color = "salmon")
plt.show()
```

#### Color Names
지원되는 140개의 색상 이름을 사용할 수 있습니다.

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "hotpink")
plt.show()
```

```python
# Hex color
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "#4CAF50")
plt.show()
```

#### Bar Width
bar()는 키워드 인수 width를 사용하여 막대의 너비를 설정합니다.

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.5)
plt.show()
```

#### Bar Height
barh()는 키워드 인수 height를 사용하여 막대의 높이를 설정합니다.

```python
x = np.array(["A", "B", "C", "D", "E", "F"])
y = np.array([3, 8, 1, 10, 5, 7])

plt.barh(x, y, height = 0.5)
plt.show()

# default height값은 0.8
```

### Histograms
히스토그램은 빈도 분포를 보여주는 그래프입니다.

각 주어진 간격 내에서 관측된 횟수를 나타내는 그래프입니다.

#### Create Histogram
Matplotlib에서는 hist() 함수를 사용하여 히스토그램을 만듭니다.

hist() 함수는 숫자 배열을 사용하여 히스토그램을 만들고 배열은 인수로 함수에 전송됩니다.

단순화를 위해 NumPy를 사용하여 250개의 값이 있는 배열을 무작위로 생성합니다. 여기서 값은 약 170에 집중되고 표준 편차는 10입니다.

```python
x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show()
```

### Pie Charts
#### Creating Pie Charts
Pyplot에서는 파이() 함수를 사용하여 원형 차트를 그릴 수 있습니다.

```python
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

보시다시피 원형 차트는 배열의 각 값(이 경우 [35, 25, 25, 15])에 대해 하나의 조각(쐐기(wedge)라고 함)을 그립니다.

기본적으로 첫 번째 쐐기의 플로팅은 x축에서 시작하여 시계 반대 방향으로 이동합니다.

각 쐐기의 크기는 다음 공식을 사용하여 값을 다른 모든 값과 비교하여 결정됩니다.

모든 값의 합으로 나눈 값: x/sum(x)

#### Labels
레이블 매개변수를 사용하여 원형 차트에 레이블을 추가합니다.

레이블 매개변수는 각 웨지에 대해 하나의 레이블이 있는 배열이어야 합니다.

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

#### Start Angle
언급한 바와 같이 기본 시작 각도는 x축에 있지만 시작 각도 매개변수를 지정하여 시작 각도를 변경할 수 있습니다.

startangle 매개변수는 도 단위의 각도로 정의되며 기본 각도는 0입니다.

```python
# Start the first wedge at 90 degrees
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

#### Explode
웨지 중 하나가 눈에 띄기를 원하십니까? explode 매개변수를 사용하면 그렇게 할 수 있습니다.

폭발 매개변수(지정된 경우 없음이 아님)는 각 쐐기에 대해 하나의 값이 있는 배열이어야 합니다.

각 값은 각 쐐기가 표시되는 중심으로부터의 거리를 나타냅니다.

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0, 0.2, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

#### Shadow
shadows 매개 변수를 True로 설정하여 원형 차트에 그림자를 추가합니다.

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```

#### Colors
색상 매개변수를 사용하여 각 웨지의 색상을 설정할 수 있습니다.

색상 매개변수(지정된 경우)는 각 웨지에 대해 하나의 값이 있는 배열이어야 합니다.

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show()
```

색상은 Hex, names, 아래 shortcuts 사용 가능

- 'r' - Red
- 'g' - Green
- 'b' - Blue
- 'c' - Cyan
- 'm' - Magenta
- 'y' - Yellow
- 'k' - Black
- 'w' - White

#### Legend
각 웨지에 대한 설명 목록을 추가하려면 legend() 함수를 사용하십시오.

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show()
```

#### Legend With Header
범례에 헤더를 추가하려면 legend 함수에 title 매개 변수를 추가하십시오.

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show()
```
